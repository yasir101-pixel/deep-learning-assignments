```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam,SGD
```

```python
# Load Data
data = pd.read_csv("/content/drive/MyDrive/PIAIC Deep Learning Assignments/ionosphere_data.csv")
data.head(4)
```

```python
#Encoding
data['label']=data['label'].map({'g':1,'b':0})
data.head(4)
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X=data.loc[:,data.columns!='label']
Y=data.loc[:,data.columns=='label']
```

```python
print(X.shape,Y.shape)
```

```python
# Split Data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.40,random_state=8)
```

```python
# Normalizing the train_data
mean = X_train.mean(axis=0)
X_train-= mean

std = X_train.std(axis=0)
X_train /= std
```

```python
X_train.head(3)
```

```python
X_train.isnull().sum()
```

```python
#Drop NaN col
X1_train=X_train.drop(columns='feature2')
X1_train.describe()
```

```python
# Normalizing the test data
mean = X_test.mean(axis=0)
X_test-= mean

std = X_test.std(axis=0)
X_test/= std
```

```python
#Drop NaN col
X1_test=X_test.drop(columns='feature2')
X1_test.describe()
```

```python
model = Sequential()
model.add(tf.keras.layers.Dense(64, activation='relu',
                                input_shape=(X1_train.shape[1],)))
model.add(tf.keras.layers.Dense(16,activation='relu'))
model.add(tf.keras.layers.Dense(8,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['acc'])
```

```python
model.fit(X1_train,Y_train,epochs=100, batch_size=16, verbose=1)
```

```python
test_rmsprop_score, test_acc_score = model.evaluate(X1_test, Y_test)
```

```python
prediction=model.predict(X1_test)
M=prediction.round()
M
```
